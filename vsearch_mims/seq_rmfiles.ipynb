```python
import os, glob
```

```python
PRJ = 'PRJEB6518'
```

```python
cdir = '/Volumes/bkup_vol001/tanamai/16S/'+PRJ
os.chdir(cdir)
```

```python
rmfiles = ['appended.fastq','unique.fastq']

## fastq.gz file ##
fastqgz = glob.glob('*.fastq.gz')
rmfiles = rmfiles+fastqgz

## merged (paird-end) ##
merged = glob.glob('*merged.fasta')
if len(merged)>0:
    rmfiles = rmfiles+merged

## trimmed (single-end) ##
trimmed = glob.glob('*trimmed.fastq')
if len(trimmed)>0:
    rmfiles = rmfiles+trimmed
```

```python
for file in rmfiles:
    print( file )
    cmd = 'rm %s' % (file)
    os.system(cmd)
```
